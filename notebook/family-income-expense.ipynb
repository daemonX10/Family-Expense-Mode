```python
def get_cols_with_zero_std_dev(df:pd.DataFrame):
    cols_to_drop = []
    numerical_columns = df.select_dtypes(include=np.number).columns
    for col in numerical_columns:
        if df[col].std() == 0:
            cols_to_drop.append(col)
    return cols_to_drop

def get_redundant_cols(df:pd.DataFrame,missing_threshold:0.5):
    cols_to_drop = []
    cols_missing_ratios = df.isna().sum().div(df.shape[0])
    col_to_drop = list(cols_missing_ratios[cols_missing_ratios>missing_threshold].index)
    return cols_to_drop
```

```python
col_to_drop1 = get_cols_with_zero_std_dev(family_income_data)
col_to_drop2 = get_redundant_cols(family_income_data,0.5)
```

```python
# Initialize an empty dictionary to store the results
results = {}

# Iterate over each column in the sample_data DataFrame
for i in sample_data.columns:
    if sample_data[i].dtype != 'object':
        # Store the results in the dictionary
        results[i] = {
            'Std Dev': int(sample_data[i].std()),
            'Mean': int(sample_data[i].mean()),
            'Median': sample_data[i].median()
        }

# Print the results dictionary
for feature, values in results.items():
    print(f"Feature: {feature}")
    for measure, value in values.items():
        print(f"{measure}: {value}")
    print("\n")
```

```python
results = {}
zero_features = []

# Iterate over each column in the family_income_data DataFrame
for i in family_income_data.columns:
    if family_income_data[i].dtype != 'object':
        # Calculate the statistics
        std_dev = int(family_income_data[i].std())
        mean = int(family_income_data[i].mean())
        median = family_income_data[i].median()

        # Store the results in the dictionary
        results[i] = {
            'Std Dev': std_dev,
            'Mean': mean,
            'Median': median
        }

        # If all statistics are zero, add the feature to the zero_features list
        if std_dev == 0 or (std_dev ==0 and mean == 0 and median == 0):
            zero_features.append(i)

# Print the results dictionary
for feature, values in results.items():
    print(f"Feature: {feature}")
    for measure, value in values.items():
        print(f"{measure}: {value}")
    print("\n")

# Print the features with all zero statistics
print("Features with all zero statistics:")
for feature in zero_features:
    print(feature)
```
